```r
# Instalar pacotes necessários
if(!require("tidyverse")) install.packages("tidyverse")
if(!require("viridis")) install.packages("viridis")

# Carregar bibliotecas
library(tidyverse)
library(viridis)

# Dataset como string
data_string <- "ID,Group,Age,Gender,Programming_Experience,Anxiety_State,Self_Efficacy,Condition_LLM,Condition_Blend,task_1,task_2,task_3,task_4,task_5,task_6,task_7,task_8,task_9,task_10,Eye_Fixation_Duration,Eye_Fixation_Count,Eye_Saccade_Amplitude,Eye_Pupil_Diameter
1,1,21,1,2,35,85,1,1,8.8,8.5,9.2,9,8.7,9.3,8.9,9.1,9.4,9.2,250,120,5.2,3.5
2,1,22,2,3,32,88,1,1,9.1,8.8,9.5,9.3,9,9.6,9.2,9.4,9.7,9.5,220,110,4.8,3.2
3,1,20,1,2,38,82,1,1,8.5,8.2,9,8.7,8.4,9.1,8.6,8.9,9.2,9,270,130,5.5,3.8
4,1,23,2,3,30,90,1,1,9.3,9,9.7,9.5,9.2,9.8,9.4,9.6,9.9,9.7,210,100,4.5,3.1
5,1,21,1,2,36,84,1,1,8.7,8.4,9.1,8.9,8.6,9.2,8.8,9,9.3,9.1,260,125,5.1,3.4
6,2,22,1,3,34,86,1,0,8.9,8.6,9.3,9.1,8.8,9.4,9,9.2,9.5,9.3,240,115,4.9,3.3
7,2,20,2,2,37,83,1,0,8.6,8.3,9.1,8.8,8.5,9.2,8.7,8.9,9.3,9.1,280,135,5.6,3.9
8,2,23,1,3,33,87,1,0,9,8.7,9.4,9.2,8.9,9.5,9.1,9.3,9.6,9.4,230,105,4.6,3.0
9,2,21,2,2,39,80,1,0,8.4,8.1,8.9,8.6,8.3,9,8.5,8.7,9.1,8.9,290,140,5.8,4.0
10,2,22,1,3,31,89,1,0,9.2,8.9,9.6,9.4,9.1,9.7,9.3,9.5,9.8,9.6,200,95,4.4,2.9
11,3,20,2,2,35,85,0,1,8.8,8.5,9.2,9,8.7,9.3,8.9,9.1,9.4,9.2,250,120,5.2,3.5
12,3,23,1,3,32,88,0,1,9.1,8.8,9.5,9.3,9,9.6,9.2,9.4,9.7,9.5,220,110,4.8,3.2
13,3,21,2,2,38,82,0,1,8.5,8.2,9,8.7,8.4,9.1,8.6,8.9,9.2,9,270,130,5.5,3.8
14,3,22,1,3,30,90,0,1,9.3,9,9.7,9.5,9.2,9.8,9.4,9.6,9.9,9.7,210,100,4.5,3.1
15,3,20,2,2,36,84,0,1,8.7,8.4,9.1,8.9,8.6,9.2,8.8,9,9.3,9.1,260,125,5.1,3.4
16,4,23,2,3,34,86,0,0,8.9,8.6,9.3,9.1,8.8,9.4,9,9.2,9.5,9.3,240,115,4.9,3.3
17,4,21,1,2,37,83,0,0,8.6,8.3,9.1,8.8,8.5,9.2,8.7,8.9,9.3,9.1,280,135,5.6,3.9
18,4,22,2,3,33,87,0,0,9,8.7,9.4,9.2,8.9,9.5,9.1,9.3,9.6,9.4,230,105,4.6,3.0
19,4,20,1,2,39,80,0,0,8.4,8.1,8.9,8.6,8.3,9,8.5,8.7,9.1,8.9,290,140,5.8,4.0
20,4,23,2,3,31,89,0,0,9.2,8.9,9.6,9.4,9.1,9.7,9.3,9.5,9.8,9.6,200,95,4.4,2.9
21,1,20,1,1,37,83,1,1,8.6,8.3,9,8.8,8.5,9.2,8.7,8.9,9.2,9,265,128,5.3,3.6
22,2,22,2,2,35,86,1,0,8.9,8.6,9.3,9.1,8.8,9.4,9,9.2,9.5,9.3,275,132,5.4,3.7
23,3,21,1,3,33,89,0,1,9.2,8.9,9.6,9.4,9.1,9.7,9.3,9.5,9.8,9.6,235,108,4.7,3.1
24,4,23,2,1,38,80,0,0,8.4,8.1,8.9,8.6,8.3,9,8.5,8.7,9.1,8.9,285,137,5.7,3.9
25,1,19,2,2,36,84,1,1,8.7,8.4,9.1,8.9,8.6,9.2,8.8,9,9.3,9.1,255,123,5.0,3.4
26,2,24,1,3,34,87,1,0,9,8.7,9.4,9.2,8.9,9.5,9.1,9.3,9.6,9.4,245,118,4.9,3.3
27,3,22,2,1,37,82,0,1,8.5,8.2,9,8.7,8.4,9.1,8.6,8.9,9.2,9,265,128,5.3,3.6
28,4,20,1,2,35,85,0,0,8.8,8.5,9.2,9,8.7,9.3,8.9,9.1,9.4,9.2,255,123,5.0,3.4
29,1,23,2,3,32,88,1,1,9.1,8.8,9.5,9.3,9,9.6,9.2,9.4,9.7,9.5,225,112,4.8,3.2
30,2,21,1,2,38,82,1,0,8.5,8.2,9,8.7,8.4,9.1,8.6,8.9,9.2,9,275,132,5.5,3.8
31,3,20,2,3,30,90,0,1,9.3,9,9.7,9.5,9.2,9.8,9.4,9.6,9.9,9.7,215,102,4.6,3.1
32,4,22,1,1,39,78,0,0,8.2,7.9,8.8,8.5,8.1,8.9,8.4,8.7,9,8.8,295,142,5.9,4.1
33,1,21,1,2,40,80,1,1,8.4,8.1,9,8.6,8.2,9,8.5,8.8,9.1,8.9,260,125,5.1,3.5
34,2,23,2,3,38,83,1,0,8.7,8.4,9.2,8.9,8.6,9.3,8.8,9.1,9.4,9.2,250,120,5.2,3.6
35,3,20,1,1,36,81,0,1,8.3,8,8.9,8.6,8.3,9,8.5,8.8,9.1,8.9,270,130,5.4,3.7
36,4,22,2,2,34,84,0,0,8.6,8.3,9.1,8.8,8.5,9.2,8.7,8.9,9.3,9.1,240,115,4.9,3.3
37,1,19,2,2,37,79,1,1,8.1,7.8,8.7,8.4,8,8.8,8.3,8.6,8.9,8.7,280,135,5.6,3.9
38,2,24,1,3,35,86,1,0,8.9,8.6,9.3,9.1,8.8,9.4,9,9.2,9.5,9.3,230,105,4.7,3.1
39,3,22,2,1,33,83,0,1,8.6,8.3,9.1,8.8,8.5,9.2,8.7,8.9,9.3,9.1,290,140,5.8,4.0
40,4,20,1,2,38,80,0,0,8.4,8.1,8.9,8.6,8.3,9,8.5,8.7,9.1,8.9,200,95,4.4,2.9"

# Ler o dataset
df <- read.csv(text = data_string)

# Normalizar variáveis neurofisiológicas para evitar escalas desiguais
df_normalized <- df %>%
  mutate_at(vars(Eye_Fixation_Duration, Eye_Fixation_Count, Eye_Saccade_Amplitude, Eye_Pupil_Diameter), scale)

# Adicionar rótulos de condições experimentais
df_normalized$LLM_Label <- factor(ifelse(df$Condition_LLM == 1, "Com LLM", "Sem LLM"))
df_normalized$Blend_Label <- factor(ifelse(df$Condition_Blend == 1, "Com Blend", "Sem Blend"))

# Definir um tema alternativo para os gráficos
theme_set(theme_minimal() + theme(plot.background = element_rect(fill = "#f0f0f0")))

# Função para gerar gráficos de densidade para variáveis neurofisiológicas
gerar_grafico_densidade <- function(var1, var2) {
  ggplot(df_normalized, aes(x = .data[[var1]], fill = interaction(LLM_Label, Blend_Label))) +
    geom_density(alpha = 0.7) +
    scale_fill_viridis(discrete = TRUE, option = "magma") +
    labs(title = paste("Comparação de", var1, "e", var2),
         x = var1,
         y = "Densidade") +
    theme(legend.position = "bottom", legend.title = element_blank())
}

# Função para gerar boxplots comparando neurofisiologia e escalas
gerar_grafico_caixa <- function(var1, var2) {
  ggplot(df_normalized, aes(x = interaction(LLM_Label, Blend_Label), y = .data[[var1]], fill = interaction(LLM_Label, Blend_Label))) +
    geom_boxplot(alpha = 0.7) +
    scale_fill_viridis(discrete = TRUE, option = "magma") +
    labs(title = paste("Boxplot de", var1, "e", var2),
         x = "Condição Experimental",
         y = var1) +
    theme(legend.position = "bottom", legend.title = element_blank())
}

# Geração de gráficos: comparações de duas escalas ou neurofisiologia
variables <- c("Anxiety_State", "Self_Efficacy", "Eye_Fixation_Duration", "Eye_Fixation_Count", "Eye_Saccade_Amplitude", "Eye_Pupil_Diameter")

graficos_densidade <- list()
graficos_caixa <- list()

# Gerar gráficos de densidade e boxplots para pares de variáveis
for (i in 1:(length(variables)-1)) {
  for (j in (i+1):length(variables)) {
    graficos_densidade[[paste(variables[i], variables[j], sep = "vs")]] <- gerar_grafico_densidade(variables[i], variables[j])
    graficos_caixa[[paste(variables[i], variables[j], sep = "vs")]] <- gerar_grafico_caixa(variables[i], variables[j])
  }
}

# Exibir os gráficos
print(graficos_densidade)
print(graficos_caixa)
```
